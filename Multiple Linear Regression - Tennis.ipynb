```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('tennis_stats.csv')
```

```python
df.head()
```

```python
plt.scatter(df['BreakPointsOpportunities'], df['Winnings'], alpha=0.5)
plt.show()
plt.clf()
```

```python
x1 = df[['BreakPointsOpportunities']]
y1 = df[['Winnings']]
```

```python
x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, train_size=0.8)
model = LinearRegression()
model.fit(x1_train, y1_train)
score = model.score(x1_test, y1_test)
```

```python
print(f'Model Score using BreakPointsOpportunities: {score}')
```

```python
y_pred1 = model.predict(x1_test)
```

```python
plt.scatter(y1_test, y_pred1, alpha=0.4)
plt.show()
plt.clf()
```

```python
x2 = df[['BreakPointsOpportunities', 'FirstServeReturnPointsWon']]
y2 = df[['Winnings']]
```

```python
x2_train, x2_test, y2_train, y2_test = train_test_split(x2, y2, train_size=0.8)
model2 = LinearRegression()
model2.fit(x2_train, y2_train)
score2 = model2.score(x2_test, y2_test)
```

```python
print(f'Model Score using BreakPointsOpportunities and FirstServeReturnPointsWon: {score2}')
```

```python
y_pred2 = model2.predict(x2_test)
```

```python
plt.scatter(y2_test, y_pred2, alpha=0.4)
plt.show()
plt.clf()
```
